## Préparation des données

Cette étape a pour but de préparer les données pour l'entraînement de notre modèle LUIS en extrayant les informations pertinentes des dialogues. Il a été conçus de sorte à être performant et lisible afin de faciliter sa compréhension et son utilisation par l'équipe de Fly Me.

Il s'agit d'une étape cruciale dans la création de notre modèle LUIS, car ces informations extraites seront utilisées pour entraîner notre modèle afin qu'il soit capable de comprendre les intentions des utilisateurs dans les dialogues.

```python
def parse_data_luis(data: pd.DataFrame) -> pd.DataFrame:
    """
    Cette fonction prend en entrée un DataFrame contenant les données de
    dialogue et extrait les informations de l'intention, la ville de départ, 
    la ville de destination, la date de départ, la date de fin et le budget. 
    Les informations extraites sont ensuite stockées dans un nouveau DataFrame
    qui est retourné en sortie.

    :param df_data: Un DataFrame contenant les données de dialogue
    :return: Un DataFrame contenant les informations extraites
    """
    col_names = [
        "text",
        "intent",
        "or_city",
        "dst_city",
        "str_date",
        "end_date",
        "n_adults",
        "n_children",
        "budget",
    ]
    # Créer un DataFrame vide avec les colonnes nécessaires
    df_parsed = pd.DataFrame(columns=col_names)
    # Utiliser une compréhension de liste pour itérer sur les tours de
    # dialogue
    for turn in data["turns"]:
        text = turn[0]["text"]
        acts = turn[0]["labels"]["acts"]
        # Utiliser des opérateurs ternaires pour extraire l'intention
        intent = (
            acts[0]["name"] if acts and acts[0]["name"] == "greeting" else
            acts[0]["args"][0]["val"] if acts and acts[0]["args"] else "")
        # Utiliser une compréhension de liste pour itérer sur les arguments de
        # chaque action de dialogue
        for act in acts[1:]:
            for arg in act["args"]:
                # Ajouter les informations extraites au DataFrame
                if arg["key"] in col_names:
                    df_parsed.loc[len(df_parsed)] = [
                        text,
                        intent,
                        *[
                            arg.get("val", "") if arg["key"] == col else ""
                            for col in col_names[2:]
                        ],
                    ]
    return df_parsed
```

## Nettoyage des données

Nous allons nettoyer les données extraites avant de les utiliser pour entraîner un modèle de compréhension de la langue naturelle (NLP) tel que LUIS. Nous allons effectuer différentes opérations de nettoyage sur les données, telles que la suppression de valeurs non numériques dans la colonne 'budget' et la conversion de cette colonne en type numérique (float).

Il est important de nettoyer les données avant de les utiliser pour entraîner un modèle de NLP car des données incomplètes ou mal formatées peuvent entraîner des résultats inattendus ou imprécis. Le nettoyage des données garantit également que les informations les plus pertinentes sont utilisées pour entraîner le modèle, ce qui améliore la qualité de la compréhension de la langue naturelle.

```python
def extract_date(text: str) -> str:
    """
    Cette fonction prend en entrée une chaîne de caractères qui représente un
    dialogue et utilise la librairie spacy pour extraire la date contenue dans
    cette chaîne de caractères.
    Elle retourne la date extraite sous forme de datetime.date ou None si
    aucune date n'a été trouvée.

    :param text: une chaîne de caractères contenant un dialogue
    :return: datetime.date ou None
    """
    nlp = spacy.load("en_core_web_lg", disable=["parser", "ner"])
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == "DATE":
            try:
                return datetime.strptime(ent.text, "%Y-%m-%d").date()
            except ValueError:
                return None
    return None


def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Cette fonction prend en entrée un DataFrame contenant les données de
    dialogue, utilise la fonction extract_date pour extraire les dates de
    départ et de fin contenues dans les dialogues,
    convertit le budget en float, et remplace les valeurs manquantes ou nulles
    par des chaînes vides. Elle retourne le DataFrame nettoyé.
    Elle utilise la librairie tqdm pour afficher une barre de progression
    détaillée.

    :param df: Un DataFrame contenant les données de dialogue
    :return: Un DataFrame nettoyé
    """
    #     print("Début de l'extraction des dates de départ")
    #     df['str_date'] = [
    #         date for date in tqdm(df["str_date"].apply(extract_date),
    #                               'Extraction de str_date')
    #     ]
    #     print("Début de l'extraction des dates de fin")
    #     df['end_date'] = [
    #         date for date in tqdm(df["end_date"].apply(extract_date),
    #                               'Extraction de end_date')
    #     ]
    df["budget"] = pd.to_numeric(df["budget"], errors="coerce")
    df = df.replace({-1: "", np.nan: ""})
    return df
```

## Mise en place des intents, entités et phrases d'entraînement pour l'application LUIS

Nous allons configurer l'application LUIS en ajoutant des intents, des entités et des phrases d'entraînement. Les intents représentent les actions que l'utilisateur souhaite effectuer, les entités sont les informations spécifiques que l'utilisateur fournit pour accomplir ces actions, et les phrases d'entraînement sont utilisées pour former l'application à reconnaître les intents et les entités. En mettant en place ces éléments de base, nous serons en mesure de créer un système de compréhension de la langue naturelle efficace pour notre projet Fly Me.

```python
def get_label(utterance: str, entity_name: str, value: str) -> dict:
    """
    Cette fonction prend en entrée une phrase, le nom de l'entité et la valeur
    de l'entité, et renvoie un dictionnaire contenant le nom de l'entité,
    l'index de début et l'index de fin de la valeur dans la phrase.
    """
    try:
        value = str(value)
        start = utterance.index(value)
        return {
            "entity_name": entity_name,
            "start_char_index": start,
            "end_char_index": start + len(value),
        }
    except (ValueError, TypeError):
        raise ValueError(
            f"La valeur {value} n'a pas été trouvée dans l'utterance {utterance}"
        )
```

### Création d'intent et d'entités sur l'application LUIS

Nous allons découvrir comment créer un intent 'OrderTravelIntent' sur l'application LUIS et ajouter des entités telles que 'DepartureCity', 'ArrivalCity', 'DepartureDate', 'ArrivalDate', 'Price' à cet intent. Nous allons également voir comment ajouter des exemples d'utterances avec les labels correspondants aux entités dans cet intent pour entraîner notre modèle.

```python
intent_name_order_travel = "OrderTravelIntent"


def add_order_travel_intent(client, version_id, app_id) -> str:
    """
    Cette fonction permet de créer un intent nommé "OrderTravelIntent" pour
    l'application LUIS en utilisant le client fourni, avec l'ID de version et
    l'ID de l'application donnés. Elle renvoie l'ID de l'intent créé.
    """
    intent_id = client.model.add_intent(app_id, version_id,
                                        intent_name_order_travel)
    print(
        f"Intent {intent_name_order_travel} avec l'id {intent_id} créée avec succès."
    )
    return intent_id


def add_order_travel_entities(client, version_id, app_id) -> dict[str, str]:
    """
    Cette fonction permet d'ajouter des entités prédéfinies et personnalisées
    pour l'application LUIS en utilisant le client fourni, avec l'ID de
    version et l'ID de l'application donnés. Elle renvoie un dictionnaire
    contenant le nom de chaque entité et son ID.
    """
    client.model.add_prebuilt(app_id,
                              version_id,
                              prebuilt_extractor_names=["datetimeV2"])
    entity_names = [
        "DepartureCity",
        "ArrivalCity",
        "DepartureDate",
        "ArrivalDate",
        "Price",
    ]
    entity_ids = {}
    for entity_name in entity_names:
        entity_id = client.model.add_entity(app_id,
                                            version_id,
                                            name=entity_name)
        entity_ids[entity_name] = entity_id
        print(f"Entité {entity_name} avec l'id {entity_id} créée avec succès.")
    return entity_ids


def add_utterances_order_travel(
    client,
    version_id,
    app_id,
    df_book_intent_train: pd.DataFrame,
    nb_utterances_to_add: int,
):
    """
    Cette fonction permet d'ajouter des phrases d'entraînement pour l'intent
    "OrderTravelIntent" de l'application LUIS en utilisant le client fourni,
    avec l'ID de version et l'ID de l'application donnés, ainsi qu'un
    DataFrame contenant les phrases à ajouter. Elle ne renvoie rien.
    """
    # On divise les phrases à ajouter en batch de 100 pour éviter les erreurs
    # de dépassement de limite
    nb_batch = math.ceil(nb_utterances_to_add / 100)
    for i in range(nb_batch):
        start = i * 100
        end = min((i + 1) * 100, nb_utterances_to_add)
        utterances = []
        for i in range(start, end):
            text = df_book_intent_train.iloc[i]["text"]
            if not isinstance(text, str):
                continue
            entity_labels = []
            for entity_name, value in df_book_intent_train.iloc[i][2:].items():
                if value:
                    try:
                        entity_labels.append(
                            get_label(text, entity_name, value))
                    except ValueError as e:
                        print(e)
            utterances.append({
                "text": text,
                "intentName": intent_name_order_travel,
                "entityLabels": entity_labels,
            })
        client.examples.batch(app_id, version_id, utterances)
    print(
        f"{nb_utterances_to_add} phrases d'entraînement ont été ajoutées avec\
         succès pour l'intent {intent_name_order_travel}.")
```

```python
# Création de l'intent "OrderTravelIntent"
intent_id = add_order_travel_intent(client, version_id, app_id)

# Ajout des entités prédéfinies et personnalisées
entity_ids = add_order_travel_entities(client, version_id, app_id)

# Ajout des phrases d'entraînement pour l'intent "OrderTravelIntent"
add_utterances_order_travel(client, version_id, app_id, df_book_intent_train,
                            len(df_book_intent_train))

print(f"L'intent {intent_name_order_travel} avec l'id {intent_id} a été créé \
    avec succès avec les entités {entity_ids} et les phrases d'entraînement.")
```

### Ajout de phrases d'entraînement pour l'intention de salutations

Dans cette section, nous allons ajouter des phrases d'entraînement pour l'intention de salutations en utilisant les fonctions add_greetings_intent et add_utterances_greetings. Ces phrases d'entraînement seront utilisées pour entraîner notre modèle LUIS afin qu'il puisse comprendre les salutations de l'utilisateur et y répondre de manière appropriée dans notre application Fly Me.

```python
intent_name_greetings = "GreetingsIntent"


def add_greetings_intent(client, version_id, app_id):
    """
    Cette fonction permet d'ajouter l'intention 'GreetingsIntent'
    l'application LUIS en utilisant le client fourni, avec l'ID de version et
    l'ID de l'application donnés. Elle renvoie l'ID de l'intention créée.
    """
    intent_id = client.model.add_intent(app_id, version_id,
                                        intent_name_greetings)
    print(f"Intent {intent_name_greetings} avec l'id {intent_id} créée avec \
        succès.")
    return intent_id


def add_utterances_greetings(client, version_id, app_id,
                             df_greeting_intent_train: pd.DataFrame):
    """
    Cette fonction permet d'ajouter des phrases d'entraînement pour l'intent
    "GreetingsIntent" de l'application LUIS en utilisant le client fourni,
    avec l'ID de version et l'ID de l'application donnés, ainsi qu'un
    DataFrame contenant les phrases à ajouter. Elle affiche le nombre
    d'utterances ajoutées et renvoie le résultat de l'ajout des utterances.
    """
    # On récupère uniquement les phrases du DataFrame
    utterances_greetings = df_greeting_intent_train["text"].tolist()
    # On ajoute l'intent_name aux phrases
    utterances_greetings = [{
        "text": text,
        "intent_name": intent_name_greetings
    } for text in utterances_greetings]
    # On divise les phrases à ajouter en batch de 100 pour éviter les erreurs
    # de dépassement de limite
    nb_batch = math.ceil(len(utterances_greetings) / 100)
    for i in range(nb_batch):
        start = i * 100
        end = min((i + 1) * 100, len(utterances_greetings))
        batch_utterances = utterances_greetings[start:end]
        client.examples.batch(app_id, version_id, batch_utterances)
    print(
        f"{len(utterances_greetings)} phrases d'entraînement ont été ajoutées\
         avec succès pour l'intent {intent_name_greetings}.")
```

```python
# Ajout des exemples supplémentaires au DataFrame
def add_none_intent_examples(df: pd.DataFrame, examples: list):
    for example in examples:
        df = df.append({"text": example, "intent": "none"}, ignore_index=True)
    return df
```

```python
def add_none_intent_additional_examples(df_none_intent, added_none_intents):
    """
    Ajoute des exemples supplémentaires pour l'intention None dans un
    DataFrame existant.

    Parameters:
        - df_luis_none_intent (pandas.DataFrame): DataFrame contenant les
          exemples existants pour l'intention None.
        - added_none_intents (List[str]): Liste des exemples supplémentaires
          ajouter pour l'intention None.

    Returns:
        pandas.DataFrame : DataFrame contenant les exemples existants et les
        exemples supplémentaires pour l'intention None.
    """
    new_data = {
        "text": added_none_intents,
        "intent": "none",
        "or_city": "",
        "dst_city": "",
        "str_date": "",
        "end_date": "",
        "budget": "",
    }
    df_new_data = pd.DataFrame(data=new_data)
    df_none_intent = pd.concat([df_none_intent, df_new_data],
                               ignore_index=True)
    return df_none_intent
```

## Entraînement du modèle avec Azure

Nous allons entraîner notre modèle en utilisant la fonction train_model(). Cette fonction prend en entrée l'objet client de l'API LUIS, l'identifiant de la version de l'application et l'identifiant de l'application. Elle déclenche l'entraînement de notre modèle en utilisant la méthode train_version() de l'objet client.

La fonction vérifie ensuite régulièrement l'état de l'entraînement jusqu'à ce qu'il soit terminé. Une fois l'entraînement terminé, un message indique que notre modèle est entraîné et prêt à être testé sur le portail LUIS.

```python
def train_model(client, version_id, app_id):
    """
    Entraîne le modèle de l'application LUIS.
    Utilise la fonction train_version() de l'API LUIS pour démarrer
    l'entraînement. Vérifie régulièrement l'état d'entraînement jusqu'à ce
    qu'il soit terminé.

    Parameters:
    - client (luis.Client): Objet client de l'API LUIS
    - version_id (str): Identifiant de la version de l'application à entraîner
    - app_id (str): Identifiant de l'application à entraîner
    """
    print("Entraînement de l'application en cours...")
    async_training = client.train.train_version(app_id, version_id)
    is_trained = async_training.status == "UpToDate"

    trained_status = ["UpToDate", "Success"]
    while not is_trained:
        time.sleep(1)
        status = client.train.get_status(app_id, version_id)
        is_trained = all(m.details.status in trained_status for m in status)

    print(
        "L'application est entraînée. Vous pouvez maintenant la tester sur le portail LUIS !"
    )
```

### Création d'un client d'exécution pour évaluer le modèle

Dans cette étape, nous utilisons les informations de clé de prédiction et d'extrémité de prédiction pour créer un client d'exécution LUIS. Ce client nous permet d'interagir avec l'application LUIS publiée pour évaluer les performances de notre modèle en utilisant des phrases de test réelles. La fonction create_luis_client_runtime() prend en charge la création de ce client en utilisant les bibliothèques CognitiveServicesCredentials et LUISRuntimeClient de Azure. Le client créé est ensuite stocké dans la variable client_runtime.

```python
def init_luis_runtime_client(prediction_key: str,
                             prediction_endpoint: str) -> LUISRuntimeClient:
    """
    Initialise un client pour l'API LUIS Runtime avec les clés de prédiction
    et l'URL de l'API.

    Parameters:
    - prediction_key (str): Clé de prédiction pour l'accès à l'API
    - prediction_endpoint (str): URL de l'API LUIS Runtime

    Returns:
    - LUISRuntimeClient : Objet client pour l'API LUIS Runtime
    """
    runtime_credentials = CognitiveServicesCredentials(prediction_key)
    client_runtime = LUISRuntimeClient(endpoint=prediction_endpoint,
                                       credentials=runtime_credentials)
    return client_runtime
```

### Évaluation du modèle LUIS

Dans cette section, nous allons évaluer le modèle LUIS créé précédemment en utilisant un jeu de données de test. Pour cela, nous allons créer un client de l'application (LUISRuntimeClient) pour pouvoir envoyer des requêtes de prédiction et obtenir les résultats.

Nous utilisons ensuite la fonction evaluate_model() pour calculer différents scores tels que la précision, le rappel et le score F1 pour les intentions. Nous calculons également la précision pour les entités détectées par le modèle. Les résultats sont affichés à l'écran et un graphique des scores est également affiché pour une visualisation facile.

Il est important de noter que pour des raisons de limitation mais aussi de performance, le nombre de données utilisées pour l'évaluation sera limité à 100. Cela nous permet de vérifier rapidement la précision de notre modèle sans avoir à utiliser un grand nombre de données, ce qui pourrait ralentir le processus d'évaluation. Toutefois, cela peut avoir un impact sur la précision des résultats obtenus.

```python
def predict_intent(text: str, client_runtime, app_id: str) -> str:
    """
    Prédiction de l'intention d'une phrase en utilisant l'API LUISRuntime.

    Parameters:
    - text (str): Phrase à prédire
    - client_runtime (luis.LUISRuntimeClient): Objet client de l'API
    LUISRuntime
    - app_id (str): Identifiant de l'application à utiliser pour la prédiction

    Returns:
    - str: Nom de l'intention prédite
    """
    try:
        prediction_response = client_runtime.prediction.get_slot_prediction(
            app_id, "Production", {"query": text})
        return prediction_response.prediction.top_intent
    except Exception as e:
        print(f"Une erreur s'est produite : {e}")
        return None


def predict_entities(text: str, client_runtime, app_id: str) -> dict:
    """
    Utilise la fonction get_slot_prediction de l'API LUIS pour obtenir les
    entités prédites pour un texte donné.

    Parameters:
    - text (str): Texte pour lequel les entités doivent être détectées
    - client_runtime (luis.LUISRuntimeClient): Objet client de l'API LUIS pour
    les prédictions en production
    - app_id (str): Identifiant de l'application LUIS pour laquelle les
    entités doivent être détectées

    Returns:
    - Dictionnaire contenant les entités détectées et leurs valeurs
    """
    try:
        prediction_response = client_runtime.prediction.get_slot_prediction(
            app_id, "Production", {"query": text})
        return prediction_response.prediction.entities
    except Exception as e:
        print(f"Une erreur s'est produite : {e}")
        return {}


def evaluate_model(df_test, client_runtime, app_id, test_size=100):
    """
    Cette fonction évalue les performances du modèle en comparant les intents
    et les entités prédites à celles du jeu de test.
    Elle prend en entrée :
    - df_test : un DataFrame contenant les intents et les textes à tester
    - client_runtime : un objet de type ClientRuntime qui permet de faire des
    requêtes à l'API LUIS
    - app_id : l'ID de l'application LUIS utilisée pour les prédictions
    - test_size (optionnel) : le nombre de textes à tester (par défaut 100)
    Elle renvoie un dictionnaire contenant les scores de précision, d'accuracy
    des entités et d'accuracy
    """

    # On récupère les intents et les textes du jeu de test, limité à test_size
    # éléments
    y_true = df_test["intent"].values[:test_size]
    texts = df_test["text"].values[:test_size]
    predictions = [
        predict_intent(text, client_runtime, app_id) for text in texts
    ]

    # On fait les prédictions d'intent et d'entités en utilisant les textes
    # une seule fois
    entities_true = [
        predict_entities(text, client_runtime, app_id) for text in texts
    ]
    entities_pred = [
        predict_entities(text, client_runtime, app_id) for text in texts
    ]

    # On calcule les scores de précision, d'accuracy des entités
    precision = precision_score(y_true,
                                predictions,
                                average="weighted",
                                zero_division=1)
    entities_accuracy = accuracy_score([str(x) for x in entities_true],
                                       [str(x) for x in entities_pred])

    # On affiche les scores
    print(f"Precision: {precision:.2f}")
    print(f"Accuracy des entités: {entities_accuracy:.2f}")

    # On renvoie les scores dans un dictionnaire
    scores = {
        "precision": precision,
        "entities_accuracy": entities_accuracy,
    }
    time.sleep(5)
    return scores


def plot_scores(scores: dict) -> None:
    """
    Affiche les scores obtenus à l'évaluation du modèle sous forme de barres.
    :param scores: Dictionnaire contenant les scores obtenus à l'évaluation du
    modèle
    :return: None
    """
    scores = {k: v for k, v in scores.items() if v is not None}
    x = [i for i, key in enumerate(scores.keys())]
    y = list(scores.values())
    plt.bar(x, y)
    plt.ylim(0, 1)
    plt.ylabel("Score")
    plt.xticks(x, list(scores.keys()))
    plt.show()
```

#### Évaluation des intentions et des entités

```python
scores = evaluate_model(df_book_intent_test, client_runtime, app_id)
plot_scores(scores)
```

#### Évaluation des greetings

```python
scores = evaluate_model(df_greeting_intent_test, client_runtime, app_id)
plot_scores(scores)
```
